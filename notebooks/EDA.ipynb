```python
## Dropping the Columns whose standard Deviation is 0 and thresold is more

def get_cols_with_zero_std_dev(df:pd.DataFrame):

    cols_to_drop = []
    num_cols = [col for col in df.columns if  df[col].dtype != 'O'] # numerical Columns Only 
    for col in num_cols:
        if df[col].std() == 0:
            cols_to_drop.append(col)
    return cols_to_drop

def get_redundant_cols(df:pd.DataFrame,missing_thresh = .7):
    cols_missing_ratios = df.isna().sum().div(df.shape[0])
    cols_to_drop = list(cols_missing_ratios[cols_missing_ratios > missing_thresh].index)
    return cols_to_drop        
```

```python
df.dtypes
```

```python
## Columns Missing Values More than 70%

cols_to_drop_1 = get_redundant_cols(df,missing_thresh=0.7)
cols_to_drop_1
```

```python
cols_to_drop_2 = get_cols_with_zero_std_dev(df=df)
cols_to_drop_2.append("Wafer")
cols_to_drop_2
```

```python
df.head()
```

```python
cols_to_drop = cols_to_drop_1 + cols_to_drop_2
```

```python
x,y = df.drop(cols_to_drop,axis=1), df["Good/Bad"]
```

```python
## Dependent Features Variable

print("Shape of the Feature now : ", x.shape)
x.head()
```

```python
## Independent/ Target Variable 

print("Shape of the label :", y.shape)
y.head()
```

```python
# Data Transformation

from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
# from sklearn.preprocessing import StandardScaler

imputer = KNNImputer(n_neighbors=3)
preprocessing_pipeline = Pipeline(
    steps=[('Imputer', imputer), ('Scaler', RobustScaler())]
)
preprocessing_pipeline
```

```python
# Transform Wafers Features

X_Trans = preprocessing_pipeline.fit_transform(x)
print("Shape of the Transformed Features set: ", X_Trans.shape)
X_Trans
```

```python
# Resampling
%pip install imbalanced-learn
```

```python
from imblearn.combine import SMOTETomek

X,y = X_Trans[:,:-1],y
resampler = SMOTETomek(sampling_strategy="auto")
X_res,y_res = resampler.fit_resample(X,y)
```

```python
print("Before Sampling Shape of Training Instances :", np.c_[X,y].shape)
print("After Sampling Shape of Training Instances :", np.c_[X_res,y_res].shape)
```

```python
# Target Cats after resampling 

print(np.unique(y_res))
print(f"Value Counts: -1: {len(y_res[y_res == -1])}, 1: {len(y_res[y_res == 1])}")
```
